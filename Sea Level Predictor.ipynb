```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

### Reading the epa-sea-level.csv file using pandas

```python
df = pd.read_csv('epa-sea-level.csv')

df.head()
```

### Filtering data and plotting the scatter plot

```python
# assign year series as variable x and csiro adjusted sea level as variable y
x = df['Year']
y = df['CSIRO Adjusted Sea Level']

# plotting a plt scatter plot using the filtered data
plt.scatter(x,y)

# adding labels and title in the plt figure
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sea Level')

# adding a grid on the figure
# plt.grid(True)

plt.show()
```

### Filtering data and plotting line of best fit for years between 1880 - 2050

```python
# Calculate the line of best fit using linregress
slope, intercept, r_value, p_value, std_err = linregress(x, y)

# create a range of years from 1880 - 2050
x_fit = list(range(min(x), 2051))

# create a sea level values for the y axis
y_fit = [slope * i + intercept for i in x_fit]

# plotting a plt line using filtered data
plt.plot(x_fit, y_fit, color='red')

# adding labels and title
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sea Level')

plt.show()
```

### Filtering data and plotting line of best fit for years >= 2000

```python
# creating a new df with years 2000 and above
new_df = df[df.Year >= 2000]

# filtering the year as variable x1 and csiro adjusted sea level as variable y1
x1 = new_df['Year']
y1 = new_df['CSIRO Adjusted Sea Level']

#Calculate the line of best fit using linregress
slope, intercept, r_value, p_value, std_err = linregress(x1, y1)

# Create a range of x values for the line of best fit
x_fit1 = list(range(min(x1), 2051))

# Calculate corresponding y values for the line of best fit.
y_fit = [slope * j + intercept for j in x_fit1]

# Plot the line of best fit.
plt.plot(x_fit1, y_fit, color='blue')

# Adding the labels and title
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sea Level')

plt.show()
```
